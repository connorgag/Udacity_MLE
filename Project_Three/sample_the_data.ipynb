```python
import pandas as pd
```

```python
df = pd.read_csv('behavioral_data.csv')
```

```python
df
```

```python
df.head(1000)
```

```python
df['IMONTH'] = df['IMONTH'].str[2:-1]
df['IDAY'] = df['IDAY'].str[2:-1]
df['IYEAR'] = df['IYEAR'].str[2:-1]
df['IDATE'] = df['IDATE'].str[2:-1]
```

```python
df.to_csv('behavioral_data_cleaned.csv', index=False, encoding='utf-8')
```
